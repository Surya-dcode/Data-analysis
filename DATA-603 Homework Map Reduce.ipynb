```python
import re
import sys
from itertools import groupby
import string
```

```python
def mapfunc(m_var):
    cleanup = ''.join([i for i in m_var.lower()])
    cleanup = re.sub(r'[^\w\s]', '', cleanup)
    return [cleanup,1]
def reductfun(key, values):
    counts = [x[1] for x in values]
    return [key,sum(counts)]
```

```python
with open("file1.txt", encoding = 'utf8') as f:
    words = [word for line in f for word in line.split()]
```

```python
map_result = map(mapfunc, words)
```

```python
map_result_sorted = sorted(map_result, key = lambda x : x[0])
```

```python
reduce_result = []
for k, g in groupby(map_result_sorted, key = lambda x: x[0]):
    reduce_result.append(reductfun(k, list(g)))
```

```python
print(reduce_result)
```

```python
with open('file2.txt', encoding = 'utf8') as h:
    words1 = [word for line in h for word in line.split()]
```

```python
second_map = map(mapfunc, words1)
```

```python
second_sorted = sorted (second_map, key = lambda x:x[0])
```

```python
reducer_sec = []
for l, m in groupby(second_sorted, key = lambda x:x[0]):
    reducer_sec.append(reductfun(l,list(m)))
```
